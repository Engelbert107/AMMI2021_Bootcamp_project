```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Task 1

```
data = pd.read_csv("train_data.txt", header=None, names = ['ID', 'f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8','f9', 'f10', 'f11', 'class'])
data.head(5)
```

## Task 2

```
x = data['class']
plt.hist(x, bins = [0.5,1.5,2.5], rwidth = 0.5)
plt.xticks([1,2])
plt.show()
```

## Task 3

```
def plot_rel():
  fig = plt.figure(figsize=(15,10), dpi=80)
  x = data['f1']
  y1 = data['f2']
  y2 = data['f3']
  y3 = data['f4']
  y4 = data['f5']
  y5 = data['f6']
  y6 = data['f7']
  y7 = data['f8']
  y8 = data['f9']
  y9 = data['f10']
  y10 = data['f11']
  cl = data['class']

  plt.subplot(4, 3, 1)
  scatter = plt.scatter(x, y1, c= cl, cmap='viridis', label=cl)
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 2)
  scatter = plt.scatter(x, y2, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 3)
  scatter = plt.scatter(x, y3, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 4)
  scatter = plt.scatter(x, y4, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 5)
  scatter = plt.scatter(x, y5, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 6)
  scatter = plt.scatter(x, y6, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 7)
  scatter = plt.scatter(x, y7, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 8)
  scatter = plt.scatter(x, y8, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 9)
  scatter = plt.scatter(x, y9, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.subplot(4, 3, 10)
  scatter = plt.scatter(x, y10, c=cl, cmap='viridis')
  plt.legend(*scatter.legend_elements())

  plt.show()

plot_rel() 
```
